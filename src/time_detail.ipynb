#### Time Analysis

```python
### time analysis ###
df = pd.read_excel(r"..\data\time_detail.xlsx", sheet_name = 'data_more')
df['type'] = df['type'].replace({'Human':'Non-AI', 'Human_AI':'AI Assisted'})
df = df[~df['time_taken_secs'].isna()]
# removing faulty outlier study
df = df[~(df['time_taken_secs'] == 507)]
# Sort DataFrame based on start time
df = df.sort_values(by=['updatedById', 'modality', 'type', 'updatedAt'])

# Generate task_number column
df['task_number'] = df.groupby(['updatedById', 'modality', 'type']).cumcount() + 1
df = df[df['task_number'] <= 15]

# boxplot - Human VS Human+AI
plt.figure(figsize=(8, 10))
sns.violinplot(x='type', y='time_taken_secs', data=df[df['time_taken_secs'] < 300], inner=None)
sns.boxplot(x='type', y='time_taken_secs', data=df[df['time_taken_secs'] < 300])
plt.xlabel('Study Type')
plt.ylabel('Time in Seconds')
plt.title('Time Taken for a Study - Non-AI VS AI Assisted')
plt.show()

# boxplot - Human VS Human+AI by Radiologist
plt.figure(figsize=(18, 10))
sns.boxplot(x='updatedById', y='time_taken_secs', data=df[df['time_taken_secs'] < 300])
plt.ylabel('Time')
plt.title('Time Taken For a Study - By User')
plt.show()

# boxplot - Human VS Human+AI by Radiologist and Phase type
plt.figure(figsize=(16, 6))
sns.boxplot(x='updatedById', y='time_taken_secs', hue='type', data=df)
plt.title('Time taken by User and Type')
plt.show()

# boxplot - Human VS Human+AI by Modality and Phase type
# Reduce the figure size
plt.figure(figsize=(9, 6))
# Adjust the box width to make them more compact
sns.boxplot(x='modality', y='time_taken_secs', hue='type', data=df, width=0.5, boxprops=dict(edgecolor=(0, 0, 0, 0)))
# Remove the border lines around the entire figure
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.ylim(0,300)
plt.xlabel('Modality')
plt.ylabel('Time in Seconds')
plt.rcParams.update({'font.size': 16})  # Adjust the font size as needed
plt.show()
```

```python
### TIME ANALYSIS ###
df = pd.read_excel(r"..\data\time_detail.xlsx", sheet_name = 'data_more')
df = df[~df['time_taken_secs'].isna()]

# removing faulty outlier study
df = df[~(df['time_taken_secs'] == 507)]

# filter for first 15 studies only
df = df.sort_values(by=['updatedById', 'modality', 'type', 'updatedAt'])
df['task_number'] = df.groupby(['updatedById', 'modality', 'type']).cumcount() + 1
df = df[df['task_number'] <= 15]

# Sort DataFrame based on start time
df = df.sort_values(by=['updatedById', 'modality', 'type', 'updatedAt'])
df['type'] = df['type'].replace({'Human':'Non-AI', 'Human_AI':'AI Assisted'})

# Generate task_number column
df['task_number'] = df.groupby(['updatedById', 'modality', 'type']).cumcount() + 1

# Loop through users and plot graphs
unique_users = df['updatedById'].unique()
unique_modality = df['modality'].unique()
```

```python
# boxplots by modality and study number

for mod in unique_modality:
    df_mod = df[df['modality'] == mod].sort_values(by = 'type')

    plt.figure(figsize=(12, 6))
    
    # Boxplot using seaborn
    sns.boxplot(x='task_number', y='time_taken_secs', hue='type', data=df_mod)
    
    plt.xlabel('Task Number')
    plt.xlim(-1,14)
    plt.ylim(0, 250)
    plt.ylabel('Time Taken (secs)')
    plt.title(f'{mod} Distribution of Time Taken by All Users for Each Task Number')
    plt.legend(title='Task Type')
    plt.tight_layout()
    
    plt.show()
```

```python
# overall time taken by study number

plt.figure(figsize=(12, 6))

# Boxplot using seaborn
sns.boxplot(x='task_number', y='time_taken_secs', hue='type', data=df.sort_values(by = 'type'))

plt.xlabel('Task Number')
plt.xlim(-1,14)
plt.ylim(0, 250)
plt.ylabel('Time Taken (secs)')
plt.title(f'Overall Distribution of Time Taken by All Users for Each Task Number')
plt.legend(title='Task Type')
plt.tight_layout()

plt.show()
```
